```python
import random
import csv
from helper import create_connection, execute_read_query, new_record

with create_connection("../PSAH.sqlite") as conn:
    cursor = conn.cursor()
    rows_sub = execute_read_query(cursor, "SELECT subject_name FROM subjects")
    subjects = [row[0] for row in rows_sub]

num_of_records = 25000
more_success_num = round(num_of_records/4*3)
less_success_num = num_of_records - more_success_num

with open('application data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    field = ["applied subject", "year until now", 'GPA', 'current subject', 'is honor college', 'GMAT', 'GRE',
              'english test exempted', 'TOFEL', 'IELTS', 'internship', 'result']
    writer.writerow(field)
    data = []
    for i in range(more_success_num):
        subject = random.choice(subjects)
        year_until_now = random.randint(0,10)
        GPA = round((random.random() * 1.0 + 3.0), 2)
        current_subject = random.choice([random.choice(subjects), subject, subject, subject, subject])
        honor_college = random.choice([1, 1, 1, 1, 0])
        GMAT = round((random.random() * 100.0 + 700.0), 2) # Scaled from 200 to 800
        GRE = round((random.random() * 40.0 + 300.0), 2) # Scaled from 130 to 340
        GMAT = random.choice([GMAT, 0])
        if GMAT != 0:
            GRE = random.choice([GRE, 0, 0])
        exempted = random.choice([0, 1])
        TOFEL = 0 if exempted else round((random.random() * 40.0 + 80.0), 2) # Scaled from 0 to 120
        IELTS = 0 if exempted else round((random.random() * 3.0 + 6.0), 2) # Scaled from 0 to 9
        TOFEL = random.choice([TOFEL, 0])
        if TOFEL != 0:
            IELTS = random.choice([IELTS, 0, 0, 0])
        internship = random.choice([1, 1, 1, 1, 0])
        result = random.choice([1, 1, 1, 1, 0])

        data.append([subject, year_until_now, GPA, current_subject, honor_college, GMAT, GRE, exempted, TOFEL, IELTS, internship, result])

    for i in range(less_success_num):
        subject = random.choice(subjects)
        year_until_now = random.randint(0,10)
        GPA = round((random.random() * 1.3 + 2.0), 2)
        current_subject = random.choice(subjects)
        honor_college = random.choice([0, 0, 0, 0, 1])
        GMAT = round((random.random() * 350.0 + 300.0), 2) # Scaled from 200 to 800
        GRE = round((random.random() * 100.0 + 180.0), 2) # Scaled from 130 to 340
        GMAT = random.choice([GMAT, 0])
        if GMAT != 0:
            GRE = random.choice([GRE, 0, 0, 0])
        exempted = random.choice([0, 1])
        TOFEL = 0 if exempted else round((random.random() * 40.0 + 80.0), 2) # Scaled from 0 to 120
        IELTS = 0 if exempted else round((random.random() * 3.0 + 6.0), 2) # Scaled from 0 to 9
        TOFEL = random.choice([TOFEL, 0])
        if TOFEL != 0:
            IELTS = random.choice([IELTS, 0, 0])
        internship = random.choice([1, 0, 0, 0, 0])
        result = random.choice([0, 0, 0, 1, 0])

        data.append([subject, year_until_now, GPA, current_subject, honor_college, GMAT, GRE, exempted, TOFEL, IELTS, internship, result])
    
    random.shuffle(data)
    for row in data:
        writer.writerow(row)
```
